```python
N = 1000
I0 = 1
S0 = 999
t = 0
deltat = 0.1

G = 0.5
B1 = 1
B2 = 1.5
B3 = 2
```

```python
S1 = []
I1 = []
R1 = []
time = []
```

```python
S1.append(S0)
I1.append(I0)
R1.append(0)
time.append(0)
```

```python
t = 1
while t < 1000:
    S1.append(S1[t-1] + deltat * (-B1 * S1[t-1] * I1[t-1] / N))
    I1.append(I1[t-1] + deltat * (B1 * S1[t-1] * I1[t-1] / N - G * I1[t-1]))
    R1.append(R1[t-1] + deltat * (G * I1[t-1]))
    #print(S[t])
    #print(I[t])
    #print(R[t])
    #print(t)
    time.append(t)
    t += 1
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(time, S1, label = 'S Ezra')
plt.plot(time, I1, label = 'I Ezra')
plt.plot(time, R1, label = 'R Ezra')
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('SIR vs. time for B = 1')
plt.legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as opt
```

```python
R01 = 0.9
R02 = 1.0
R03 = 1.1
R04 = 1.2
```

```python
R = np.linspace(0,1,100)
G = 1 - np.exp(-R01 * R)
F = R

def f(x):
    return x
def g(x):
    return 1 - np.exp(-R01 * x)
def findIntersection(fun1,fun2,x0):
 return opt.fsolve(lambda x : fun1(x) - fun2(x), x0)
intersection = findIntersection(f,g,0)

fig,ax = plt.subplots()
ax.plot(R, G, label='g', color = 'red')
ax.plot(R, F, label='f', color = 'black')
ax.scatter(intersection, intersection, color = 'blue')
ax.legend()
plt.title('f = g for R0 = 0.9')
ax.set_xlabel('x values') 
ax.set_ylabel('y values')
```

```python
N = 1000
I0 = 1
S0 = 999
t = 0
deltat = 0.1
G = 0.5
B1 = 1
S1 = []
I1 = []
R1 = []
time = []
S1.append(S0)
I1.append(I0)
R1.append(0)
time.append(0)
t = 1

R0 = B1 / G # 2
def f(x):
    return x
def g(x):
    return 1 - np.exp(-R0 * x)
def findIntersection(fun1,fun2,x0):
 return opt.fsolve(lambda x : fun1(x) - fun2(x), x0)
intersection = findIntersection(f,g,1.0)
r_inf = [intersection * N]

while t < 1000:
    S1.append(S1[t-1] + deltat * (-B1 * S1[t-1] * I1[t-1] / N))
    I1.append(I1[t-1] + deltat * (B1 * S1[t-1] * I1[t-1] / N - G * I1[t-1]))
    R1.append(R1[t-1] + deltat * (G * I1[t-1]))
    #print(S[t])
    #print(I[t])
    #print(R[t])
    #print(t)
    time.append(t)
    r_inf.append(intersection * N)
    t += 1 
    
plt.plot(time, S1, label = 'S Ezra')
plt.plot(time, I1, label = 'I Ezra')
plt.plot(time, R1, label = 'R Ezra')
plt.plot(time, r_inf, label = 'r_inf', linestyle = 'dashed', color = 'green')
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('SIR vs. time for B = 1 with r_inf')
plt.legend()
```

```python
N = 10 ** 6
I0 = 1
S0 = N - I0
t = 0
deltat = 0.1
G = 0.5
B1 = 1
S1 = []
I1 = []
R1 = []
time = []
S1.append(S0)
I1.append(I0)
R1.append(0)
time.append(0)
t = 1

while t < 1000:
    S1.append(S1[t-1] + deltat * (-B1 * S1[t-1] * I1[t-1] / N))
    I1.append(I1[t-1] + deltat * (B1 * S1[t-1] * I1[t-1] / N - G * I1[t-1]))
    R1.append(R1[t-1] + deltat * (G * I1[t-1]))
    time.append(t)
    t += 1 
    
I1[t-1] = I1[t-1] + 1 / N    
while t < 2000:
    S1.append(S1[t-1] + deltat * (-B1 * S1[t-1] * I1[t-1] / N))
    I1.append(I1[t-1] + deltat * (B1 * S1[t-1] * I1[t-1] / N - G * I1[t-1]))
    R1.append(R1[t-1] + deltat * (G * I1[t-1]))
    time.append(t)
    t += 1   
    
plt.plot(time, S1, label = 'S Ezra')
plt.plot(time, I1, label = 'I Ezra')
plt.plot(time, R1, label = 'R Ezra')
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('SIR vs. time with E = 1 / N perturbation at t = 1000')
plt.legend()
```
